```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cap=cv2.VideoCapture("harder_challenge_video.mp4")
# ret,frame=cap.read()
# img=np.copy(frame)
#####################################
def canny(image):
    grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    blur=cv2.GaussianBlur(grey,(7,7),0)
    canny=cv2.Canny(blur,50,150)
    
    return canny
##################################################
def ROI(image):
    black=np.zeros_like(image)
    polygon=np.array([[(150,image.shape[0]),(1090,image.shape[0]),(645,420)]])
    cv2.fillPoly(black,polygon,255)
    masked=cv2.bitwise_and(image,black)
    return masked
def hough(image,lines):
    if lines is not None:
        for lin in lines:
            x1,y1,x2,y2=lin[0]
            cv2.line(image,(x1,y1),(x2,y2),(0,255,0),5)
    return image

# cv2.imshow("img",frame)

# cann=canny(frame)
# cv2.imshow("img",cann)

# roi=ROI(cann)
# cv2.imshow("img",roi)

# lines=cv2.HoughLinesP(roi,2,np.pi/180,50,minLineLength=10, maxLineGap=100)
# h=hough(frame,lines)
# cv2.imshow("img",h)
while True:
    ret,frame=cap.read()
    cv2.imshow("img",frame)

    cann=canny(frame)
    cv2.imshow("img",cann)

    roi=ROI(cann)
    cv2.imshow("img",roi)

    lines=cv2.HoughLinesP(roi,1,np.pi/180,100,minLineLength=10, maxLineGap=50)
    h=hough(frame,lines)
    cv2.imshow("img",h)
    if cv2.waitKey(1)==13:
        break
cap.release()
# cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
